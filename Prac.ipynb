Handling Categorical variales

```python
Categorical_cols=train.select_dtypes(include='object')
Categorical_cols.head()
```

```python
print('We have {} categorical columns out of {} total columns'.format(len(Categorical_cols.columns),train.shape[1:]))
```

```python
# Create the list of features based on your intuition first
feature_names =  ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
# Select data corresponding to features in feature_names as 
A = train[feature_names]
```

```python
#Select the prediction target which is the Sale Price column
b = train.SalePrice
```

```python
A.head()
```

```python
A.describe()
```

Building Machine Learning Model

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=1)

# Fit the model
model.fit(A, b)
```

```python
predictions = model.predict(A)
print(predictions)
```

Model validation

```python
#Calculate the mean absolute error
from sklearn.metrics import mean_absolute_error
mean_absolute_error(b, predictions)
```

Split data into training and validation data, for both features and target

```python
from sklearn.model_selection import train_test_split
train_a, val_a, train_b, val_b = train_test_split(A, b, random_state = 42)
# Define model
new_model = DecisionTreeRegressor()
# Fit model
new_model.fit(train_a, train_b)
```

```python
# get predicted prices on validation data
val_predictions = new_model.predict(val_a)
print(mean_absolute_error(val_b, val_predictions))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
Log_model = LogisticRegression(random_state=1)

# Fit the model
Log_model.fit(A, b)
# get predicted prices on validation data
Log_predictions = Log_model.predict(val_a)
print(mean_absolute_error(val_b, Log_predictions))
```

```python
Rf_model = RandomForestRegressor(random_state=1)

# Fit the model
Rf_model.fit(A, b)
# get predicted prices on validation data
Rf_predictions = Rf_model.predict(val_a)
print(mean_absolute_error(val_b, Rf_predictions))
```
